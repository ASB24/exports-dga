```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('./exportaciones_por_capitulos_2017_2023.csv', encoding='latin-1', sep=';')
print(df.shape)
df.head()
```

```python
df.rename(columns={
    'Código Capítulo': 'chapter_number',
    'Capítulo Descripción': 'chapter_description',
    'Exportaciones': 'exports',
    'Mes': 'month',
    'NUM': 'month_number',
    'Año': 'year'
}, inplace=True)
df.head()
```

```python
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month_number'].astype(str), format='%Y-%m') \
    + pd.offsets.MonthEnd(1)
df.head()
```

```python
df.set_index(['date'], inplace=True)
df.drop(['month', 'month_number', 'year'], axis=1, inplace=True)
df.head()
```
